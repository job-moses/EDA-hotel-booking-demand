By examining and visualizing the data, we can uncover meaningful insights, detect anomalies, and formulate hypotheses for further investigation.

### Data Cleaning

```python
data.head()
```

**.drop_duplicate()** 

```python
# drop duplicates rows
data.drop_duplicates(inplace=True)
```

```python
# check for changes
data.duplicated().values.any()
```

**Handling missind values**

`Agent` and `company` column contains a lot of missing values let drop these columns from the dataset

```python
data.drop(['agent','company'], inplace = True, axis = 1)
```

**.dropna()**

```python
# drop missing values in country column
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

**convert reservation status date from object to datetime data type**

```python
data['reservation_status_date'] = pd.to_datetime(data['reservation_status_date'])
```

```python
data['reservation_day'] = data['reservation_status_date'].dt.day
data['reservation_month'] = data['reservation_status_date'].dt.month
data['reservation_year'] = data['reservation_status_date'].dt.year
```

**remove outlier in `adr`**

```python
# filter rows where average daily rate less than 0
data = data[data['adr']<5000] 
```

```python
data.head()
```

### Feature Engineering

Here I will create new features to enrich the data for analysis

```python
# Total number of guest per week = stay_in week_days + stay_in_week_night
data['number_of_days'] = data['stays_in_weekend_nights'] + data['stays_in_week_nights']
```

```python
# total number of guest per booking
data['total_guest_per_booking'] = data['adults'] + data['children'] + data['babies']
```

```python
data['avg_adr_per_person'] = data['adr']/data['total_guest_per_booking']
```

### Through which channels do customers make their bookings at our hotel?

### Number of Bookings Based on Hotel Type

```python
#total booking across the hotels
hotel_bookings = data['hotel'].value_counts().reset_index() 
hotel_bookings.columns = ['Hotel_Type','Total_Bookings'] #renaming the columns

# create the percentage of bookings
hotel_bookings['pct_booking'] = hotel_bookings['Total_Bookings']/hotel_bookings['Total_Bookings'].sum() * 100
hotel_bookings['pct_booking'] = round(hotel_bookings['pct_booking'],2).astype(str) + '%'

# plot of hotel type and total bookings
g = sns.barplot(data = hotel_bookings, x ='Hotel_Type', y ='Total_Bookings',)
plt.ylabel('Number of Reservation')
g.set_title('NUMBER OF RESERVATIONS BY HOTEL TYPE')
g.bar_label(container=g.containers[0], labels = hotel_bookings['pct_booking']);
```

### Distribution of Bookings Across Years and Reservartion Status

### How does the average daily rate in city hotel and resort hotel compare

```python
city_hotel_adr = city_hotel.groupby(['reservation_status_date'])[['adr']].mean() 
resort_hotel_adr = resort_hotel.groupby(['reservation_status_date'])[['adr']].mean()
```

```python
#visualization of average daily rate in city hotel and resort hotel
plt.figure(figsize = (10,5))
plt.plot(resort_hotel_adr.index, resort_hotel_adr['adr'], label = 'Resort Hotel', color = 'y')
plt.plot(city_hotel_adr.index, city_hotel_adr['adr'], label = 'City Hotel', color = 'g' )
plt.xlabel('Reservation Status Date')
plt.ylabel('Average Daily Rate')
plt.title('Averege Daily Rate in City and Resort Hotel')
plt.legend();
```

**The average daily rate in the City hotel has shown a consistent increase, while in the Resort hotel, it has exhibited fluctuations over time.**

### Conclusion and Recommendation

**Conclusion:**

The exploratory data analysis of the hotel booking dataset has provided valuable insights into customer behavior, booking trends, and potential areas of improvement. 

Firstly, it's evident that the hotels attract guests from a wide range of countries, with a significant portion of bookings originating from Portugal, Great Britain, France, and Spain. This international customer base highlights the hotels' global appeal.

Furthermore, the dominance of online bookings, accounting for 59% of the total, underscores the importance of maintaining and enhancing the online booking platform's efficiency and user-friendliness.

The City hotel outperforms the Resort hotel in terms of bookings, comprising 61.44% of the total, but it faces a relatively higher rate of cancellations at 30%. On the other hand, the Resort hotel has a lower booking count but a more favorable cancellation rate of 23%. This suggests that the City hotel may need to implement strategies to mitigate cancellations and enhance customer retention.

The analysis of average daily rates reveals differing trends between the two hotel types. The City hotel's consistent rate increase signifies the demand for its services, while the Resort hotel's rate fluctuations might indicate potential areas for pricing optimization.

Notably, months with higher average daily rates coincide with elevated booking cancellation rates. This highlights the importance of carefully aligning pricing strategies with customer expectations and value.

The extended lead time associated with cancelled bookings (105 days on average) compared to non-cancelled ones (70 days) suggests a potential relationship between longer lead times and booking cancellations. Further investigation into this correlation could guide strategies to manage and reduce cancellations.

**Recommendations:**

1. **Cancellation Analysis:** City hotel should focus on reducing its cancellation rate by understanding the reasons behind cancellations and implementing strategies to address them. This might include offering flexible booking options, providing clear cancellation policies, and enhancing customer engagement post-booking.

2. **Pricing Strategy:** The Resort hotel should explore ways to stabilize its average daily rates to improve revenue consistency. This could involve data-driven pricing adjustments based on demand patterns and customer preferences.

3. **Customer Engagement:** Both hotels should prioritize enhancing customer engagement, especially during the booking process. This can include improving the website's usability, providing helpful information, and leveraging personalized offers to increase conversion rates.

4. **Lead Time Management:** By understanding the correlation between lead time and cancellations, hotels can consider incentivizing early bookings or offering special deals to guests who book well in advance, potentially reducing cancellation rates.

5. **Meal and Room Preferences:** Understanding that guests prefer room type A and meal type BB provides an opportunity to tailor marketing efforts and enhance these offerings to meet customer preferences.

6. **Peak Months Strategy:** Hotels can implement strategies to manage demand during peak months, potentially including dynamic pricing, additional services, or managing reservations to ensure a balance between bookings and cancellations.

In conclusion, the exploratory data analysis offers valuable insights for both operational improvements and strategic decision-making. By focusing on reducing cancellations, optimizing pricing strategies, and providing a seamless booking experience, the hotels can enhance customer satisfaction, revenue generation, and overall success.
